```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
ps
```

```python
df.head()
```

```python
def datatransform(text):
# 1 convert lower case
    text=text.lower() 
# 2 divide small parts
    text=nltk.word_tokenize(text)
# 3 special characters
    b=[]
    for n in text:
        if n.isalnum():
            b.append(n)
# 4 stop words & punctuation
    text=b[:]
    b.clear()
    for n in text:
        if n not in sw and n not in pn:
            b.append(n)
# 5 staming
    text=b[:]
    b.clear()
    for n in text:
        b.append(ps.stem(n))
        
    return b
# datatransform('Ali Is my NAMW @ @$$@ is@@ I loveing to playing cricket')    
```

```python
datatransform('Ali Is my NAMW @ @$$@ is@@ I loveing to playing cricket')    
```

```python
df['transformed_data']=df['Email_Text'].apply(datatransform)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv = CountVectorizer()
tfidf=TfidfVectorizer(max_features=3000)
```

```python
# # df['transformed_data'] = df['transformed_data'].apply(lambda x: " ".join(x))
# X = tfidf.fit_transform(df['transformed_data']).toarray()
# X.shape
# X

df['transformed_data'] = df['transformed_data'].apply(lambda x: " ".join(x))

# apply TF-IDF
X = tfidf.fit_transform(df['transformed_data']).toarray()

print(X.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

#GaussianNB

```python
gnb.fit(X_train,y_train)
y_pred1 = gnb.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

#MultinomialNB

```python
mnb.fit(X_train,y_train)
y_pred2 = mnb.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
```
